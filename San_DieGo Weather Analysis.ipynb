## 2. Data Preparation
The dataset consists of daily observations of temperature, precipitation, and snow depth. To ensure data quality and analytical consistency, several preprocessing steps were performed.  
Due to a large proportion of missing values (over 78%) in the original `TAVG` column, a new variable `TAVG_computed` was created using the average of daily maximum and minimum temperatures:

$$
\text{TAVG\_computed} = \frac{\text{TMAX} + \text{TMIN}}{2}
$$

Additionally:
- Missing values in snow-related variables were filled with 0, as snowfall is negligible in San Diego.
- Rows with missing TMAX or TMIN were removed, as they accounted for fewer than 10 records out of over 30,000 observations..

#### Create an average temperature TAVG (Degrees Fahrenheit)

The original `TAVG (Degrees Fahrenheit)` column is highly incomplete, with **over 78% of the data missing** (24,297 out of ~31,670 rows).  
Because of this, any comparison between the original TAVG and `(TMAX + TMIN)/2` would only involve a small subset of the dataset and would not be statistically meaningful.  

To address this, a new column `TAVG_computed` was created by averaging `TMAX` and `TMIN`:

$$
\text{TAVG\_computed} = \frac{\text{TMAX} + \text{TMIN}}{2}
$$

This ensures full coverage, logical consistency (`TMIN \le TAVG\_computed \le TMAX`), and provides a reliable daily average temperature for analyzing San Diego weather trends.

```python
df["TAVG_computed"] = (df["TMAX (Degrees Fahrenheit)"] + df["TMIN (Degrees Fahrenheit)"]) / 2
```

Snow-related variables (SNOW and SNWD) were excluded from the analysis because they contain no meaningful variation in the dataset. Both variables have a mean, median, and maximum value of 0, indicating that snowfall and snow depth were essentially never recorded during the observation period. In contrast, precipitation (PRCP) shows measurable variability and meaningful distribution across days. Therefore, snow variables do not contribute to climate pattern analysis and were removed to focus on more informative features.

#### 3.2 Monthly Average, Maximum, and Minimum temperature

```python
# --- Compute all monthly averages in one go ---
monthly = (
    df_clean.groupby("Month")[[
        "TAVG_computed",
        "TMAX (Degrees Fahrenheit)",
        "TMIN (Degrees Fahrenheit)"
    ]]
    .mean()
    .reset_index()
)

# Rename for convenience
monthly.rename(columns={
    "TMAX (Degrees Fahrenheit)": "TMAX",
    "TMIN (Degrees Fahrenheit)": "TMIN"
}, inplace=True)

months = monthly["Month"]
tavg = monthly["TAVG_computed"]
tmax = monthly["TMAX"]
tmin = monthly["TMIN"]

# --- Smooth curves ---
from scipy.interpolate import make_interp_spline
import numpy as np

x_smooth = np.linspace(1, 12, 400)
tavg_smooth = make_interp_spline(months, tavg)(x_smooth)
tmax_smooth = make_interp_spline(months, tmax)(x_smooth)
tmin_smooth = make_interp_spline(months, tmin)(x_smooth)

plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")
ax = plt.gca()
ax.set_facecolor("#f4f6f9")

# ============================================================
# 1) GRADIENT FILL BETWEEN TMAX & TMIN
# ============================================================
gap = tmax_smooth - tmin_smooth
norm_gap = (gap - gap.min()) / (gap.max() - gap.min())

for i in range(len(x_smooth) - 1):
    color = plt.cm.coolwarm_r(norm_gap[i])
    plt.fill_between(
        x_smooth[i:i+2],
        tmax_smooth[i:i+2],
        tmin_smooth[i:i+2],
        color=color,
        alpha=0.32,
        linewidth=0
    )

# ============================================================
# 2) GRADIENT LINE FOR TAVG (add label on first segment)
# ============================================================
for i in range(len(x_smooth)-1):
    plt.plot(
        x_smooth[i:i+2],
        tavg_smooth[i:i+2],
        color=plt.cm.coolwarm(i / len(x_smooth)),
        linewidth=3,
        zorder=4,
        label="Avg Temp (TAVG)" if i == 0 else None
    )

# ============================================================
# 3) SMOOTH LINES FOR TMAX & TMIN
# ============================================================
plt.plot(x_smooth, tmax_smooth, color="#d62728", linewidth=3.5, label="Avg Max Temp")
plt.plot(x_smooth, tmin_smooth, color="#1f77b4", linewidth=3.5, label="Avg Min Temp")

# ============================================================
# 4) RAW DATA POINTS
# ============================================================
plt.scatter(months, tavg, color="black", s=50, zorder=6)
plt.scatter(months, tmax, color="#8c1b13", s=100, edgecolor="white", linewidth=1.5, zorder=6)
plt.scatter(months, tmin, color="#0b3c78", s=100, edgecolor="white", linewidth=1.5, zorder=6)

# ============================================================
# 5) HIGHLIGHT HOTTEST & COLDEST MONTHS
# ============================================================
max_idx = tmax.idxmax()
min_idx = tmin.idxmin()

plt.scatter(months[max_idx], tmax[max_idx], color="red", s=220, edgecolor="black", zorder=10)
plt.scatter(months[min_idx], tmin[min_idx], color="blue", s=220, edgecolor="black", zorder=10)

plt.text(
    months[max_idx], tmax[max_idx] - 2,
    f"Warmest Month\n{tmax[max_idx]:.1f}°F",
    fontsize=10, weight="bold", color="red"
)

plt.text(
    months[min_idx] - 0.1, tmin[min_idx] + 1,
    f"Coldest Month\n{tmin[min_idx]:.1f}°F",
    fontsize=10, weight="bold", color="blue"
)

# ============================================================
# 6) LABELS, TITLES, GRID
# ============================================================
plt.title(
    "Monthly Average, Maximum, and Minimum Temperature",
    fontsize=18, weight="bold", pad=20
)

plt.xlabel("Month", fontsize=12)
plt.ylabel("Temperature (°F)", fontsize=12)
plt.xticks(range(1, 13))

plt.grid(alpha=0.25, linestyle="--")

# Updated legend includes Avg Temp now
plt.legend(frameon=True, facecolor="white", edgecolor="gray", fontsize=10)

plt.tight_layout()
plt.show()

```

```python
# Compute yearly extreme high and extreme low
yearly_extremes = (
    df_clean.groupby("Year")[["TMAX (Degrees Fahrenheit)", "TMIN (Degrees Fahrenheit)"]]
    .agg({"TMAX (Degrees Fahrenheit)": "max",
          "TMIN (Degrees Fahrenheit)": "min"})
    .reset_index()
)

# Prepare data
years = yearly_extremes["Year"]
highs = yearly_extremes["TMAX (Degrees Fahrenheit)"]
lows  = yearly_extremes["TMIN (Degrees Fahrenheit)"]

# Smooth curves using spline interpolation
x_smooth = np.linspace(years.min(), years.max(), 300)
high_smooth = make_interp_spline(years, highs)(x_smooth)
low_smooth  = make_interp_spline(years, lows)(x_smooth)

plt.figure(figsize=(15, 8))

# --- Background style ---
sns.set_style("whitegrid")
plt.gca().set_facecolor("#f7f7f7")

# --- Shaded envelope between extremes ---
plt.fill_between(
    x_smooth, high_smooth, low_smooth,
    color="lightgray", alpha=0.35,
    label="Temperature Range"
)

# --- Smooth high line ---
plt.plot(
    x_smooth, high_smooth,
    color="#d62728", linewidth=3,
    label="Extreme High Temp"
)

# --- Smooth low line ---
plt.plot(
    x_smooth, low_smooth,
    color="#1f77b4", linewidth=3,
    label="Extreme Low Temp"
)

# --- Actual data points (subtle markers) ---
plt.scatter(years, highs, color="#8c1b13", s=40, zorder=5)
plt.scatter(years, lows,  color="#0b3c78", s=40, zorder=5)

# --- Trendlines ---
z_high = np.polyfit(years, highs, 1)
z_low  = np.polyfit(years, lows, 1)

plt.plot(
    years, np.polyval(z_high, years),
    color="#8c1b13", linestyle="--", linewidth=2,
    label="High Temp Trend"
)

plt.plot(
    years, np.polyval(z_low, years),
    color="#0b3c78", linestyle="--", linewidth=2,
    label="Low Temp Trend"
)

# --- Highlight record-breaking years ---
max_idx = highs.idxmax()
min_idx = lows.idxmin()

max_year = years[max_idx]
min_year = years[min_idx]

max_temp = highs.max()
min_temp = lows.min()

# Plot points
plt.scatter(max_year, max_temp, color="red", s=120, edgecolor="black", zorder=10)
plt.scatter(min_year, min_temp, color="blue", s=120, edgecolor="black", zorder=10)

# --- Annotate with offset + arrows (prevents overlap) ---
plt.annotate(
    f"Record High\n{max_temp:.1f}°F ({max_year})",
    xy=(max_year, max_temp),
    xytext=(max_year + 2, max_temp),
    textcoords="data",
    fontsize=11,
    weight="bold",
    color="red",
    arrowprops=dict(arrowstyle="->", color="red", lw=1.5),
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red", alpha=0.8)
)

plt.annotate(
    f"Record Low\n{min_temp:.1f}°F ({min_year})",
    xy=(min_year, min_temp),
    xytext=(min_year + 2, min_temp),
    textcoords="data",
    fontsize=11,
    weight="bold",
    color="blue",
    arrowprops=dict(arrowstyle="->", color="blue", lw=1.5),
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="blue", alpha=0.8)
)

# --- Titles & labels ---
plt.title("Historical Extreme High and Low Temperatures by Year",
          fontsize=20, weight="bold", pad=20)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Temperature (°F)", fontsize=15)

# --- ✅ ADD MORE TICKS HERE ---
ax = plt.gca()

# X-axis (more year ticks)
ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=15))
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(2))

# Y-axis (more temperature ticks)
ax.yaxis.set_major_locator(ticker.MaxNLocator(nbins=12))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(2))

# Tick appearance
ax.tick_params(axis='both', which='major', labelsize=12)
ax.tick_params(axis='both', which='minor', length=4, color='gray')

# Grid improvement
plt.grid(which='major', linestyle='--', alpha=0.25)
plt.grid(which='minor', linestyle=':', alpha=0.15)

# --- Legend ---
plt.legend(fontsize=13, frameon=True, facecolor="white", edgecolor="gray")

plt.tight_layout()
plt.show()
```
